# Prompt Engineering #

### Hvad er Prompt Engineering? ###

Prompt engineering refererer til den systematiske metode til at designe, strukture og optimere input til en sprogmodel for at styre dens output mere præcist. Teknisk er det en metode indenfor for Natural Language Processing (NLP), hvor man systematisk designer input (prompts) til en Large Language Model (LLM) for at få forudsigelise, kontekstuelle og præcise svar. 

Matematisk betegnelse som en funktion:
O = f(P, C, H, M)

Output fra modellen = 
Prompt, den specifikke inputtekst
Kontekst, informationer
Hyperparametre, temperature, top-k sampling, max tokens, osv...
Modellens, træningstilstand og vægte

I praksis betyder det, at den samme model kan give forskellive svar afhængigt af, hvordan man formulerer en prompt.

### Hvad er de forskellige typer af prompt teknikker og hvordan gør man brug af dem? ###

- Zero-shot prompting: Blot spørgsmål, uden eksempler
F.eks.: Oversæt "Hello, how are you?" til dansk

- Few-shot prompting: Spørgsmål med eksempler, for bedre format
F.eks. 

- Using delimiters, 
- Detailed, 
- Numbered steps prompting.

- Chain-of-thought prompting
Beder modellen om at tænke i flere trin, hvilket kan gøre LLMs svar mere effektiv.

Fordel: Nyttigt til logiske og matematiske problemer.

### Hvordan skriver man de forskellige prompt teknikker? ###

Eksempel:
Input: Hvad er 27 * 43? 

(For bredt spørgsmål, hvilket kun vil vise output som i nogle tilfælde kan være forkert)

Output: Løs 27 * 43 trin for trin.

(Bedre metode, da modellen vil fortælle processen fra start til slut om hvordan den finder frem til svaret "Først ganger vi 27 med 40... Derefter lægger vi til..." som sikrer et mere korrekt svar, samtidig med at man kan være kritisk og undersøge om processen har været korrekt).
