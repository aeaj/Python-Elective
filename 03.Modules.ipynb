Hvad er et modul?

- En modul er også kendt dom en Python fil (F.eks. "script.py") som indeholder en kollektion af funktioner. Man opretter en python fil ved at tilføje dem som "new file/ny fil" i explorer i navigationsbaren til venstre under introduktion. 

Vi bruger lommeregner som følgende eksempel, for at forstå oprettelse og omdøbbelse af moduler.


Hvordan opretter man en hel modul?

```python
#Lav en python fil, f.eks. calculator
#Dernæst opret diverse relevante funktioner

#Opret en python, eller notebook fil som denne 
#Dernæst kan du importer python filen ved følgende:

import calculator #Importer python filen
calculator.add(2,2) #Du kan nu beregne forskellige værdier 
#calculator.sub(4,2)
#calculator.mult(5,5)
#calculator.div(50,5)
```

Hvordan opretter man en modul med en enkelt funktion?

```python
#Man kan også oprette et modul med en enkelt funktion

from calculator import sub #Her kan du tilføje flere funktioner
sub(9,3)
```

Hvordan navngiver jeg eller forkorter navnet på en modul?

```python
#Navngiv moduler ved hjælp af "as keyword"

import calculator as calc
calc.mult(3, 3)
```

Man kan altid bruge følgende:

- "help("modules") #Oversigt over forskellige brug af moduller.
- import es #Oversigt over atributter og funktioner.


Hvordan bruger man indbyggede moduler og pakker?

```python
#Du kan se en oversigt af indbyggede moduller ved følgende:
help('modules')
```

```python
#Oprettelse vha. indbyggede moduler

import os #Importerer...?
os.mkdir('test') #Make/Create directory = Opretter en mappe
os.chdir('test') #Change directory = Tilgår mappen
with open('test.md', 'w') as l: #Opretter tekst fil med 'test.md, hvor w = write/skrive i filen og 'as l' = navngivning af mappen
    l.write('Vi tester lige tekstfilen') #Skriver noget i tekstfilen
os.mkdir('bar') #Opretter en undermappe hvor tekstfilen oprettes
```
