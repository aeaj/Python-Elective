**Hvad er Langchain?**

Open source orkestrerings Framework der gør det lettere at udvikle AI-applikationer med LLM.

Består af:
- LLM (Kan integrere forskellige LLM'er, Mistral, OpenAI, osv...)
- Prompts
- Chains, sekventiel kæde
- Indeks (Document loader, Vector Databases, Text Splitters)
- Memory (Huske tidligere samtalehistorik)
- Agents 

**Hvad kan Langchain bruges til?**
- Chatbot
- Summarization
- Question answering
- Data augmentation
- Virtuelle agents (RPA)

Langchains værktøjer og API'er forenkler processen i udvikling af applikationer der gør brug af LLM.

**Hvad er koncepterne indexing, retrieval og generation i en RAG sammenhæng?**

RAG-sammenhæng

Overordnet proces:
- Henter docs
- Split docs for embedding eller convenience
- Embed hvert split
- Gemmer det i et vektorlager (numeriske repræsentationer af docs splits, samt relevant spørgsmål i et dictionary{} som videreføres til en
- prompt template -> promptvalue -> LLM -> chatmessage to Parser -> String svar)

Question -> Indeks(Docs) -> Retrieval(Relevant Docs) -> Generation() -> Answer

PART 1 - Grundlæggende
Kodestruktur:

- Set necessary api keys for Langsmith
- Install a few packages
- Import necessary libraries and frameworks
- Set up environment variables

Indeks
- Load documents
- Split: Split the docs and set a chunk size of x characters
- Embed: Hvert split er indlejret/embedded  og indekseret i en vektor-store (F.eks. Chroma lokalt)

Retrieval and Generation
- Prompt: Definer en prompt for RAG
- LLM: Find en relevant LLM (F.eks. OpenAI gpt3.5)
- Post-Process: Definerer docs og returnerer kontekst
- Chain: Tager input spørgsmål, kør retrieveren til at afhent/fetch docs, indsæt de afhentet/retrieved docs og bruger spørgsmål/userquery til vores prompt, overfører det til LLM, formaterer output som en string og hermed får vi vist den endelig output for modellen.
